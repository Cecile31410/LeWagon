```python
from google.colab import auth
import pandas as pd
auth.authenticate_user()
query = "SELECT * FROM team-elec.Merge_Projet.Projet_Elec_per_hab"
df_hab = pd.read_gbq(query, project_id = "team-elec")
```

```python
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_hab.head()
```

```python
df_hab['date_date'] = pd.to_datetime(df_hab['date_date'])

df_hab['date_date'] = df_hab['date_date'].dt.year
df_hab
```

```python
df_hab_2022 = df_hab[df_hab['date_date'] == 2022]
df_hab_2022
```

```python
df_hab_2013 = df_hab[df_hab['date_date'] == 2013]
df_hab_2013
```

```python
from google.colab import files
```

```python
row = df_hab_2022.iloc[0]


labels = ['Nuclaire', 'Hydraulique', 'olien', 'Solaire', 'Thermique', 'Export']
values = [
    row['Pct_nucl_prod_nette'],
    row['Pct_hydro_prod_nette'],
    row['Pct_eol_prod_nette'],
    row['Pct_sol_prod_nette'],
    row['Pct_therm_prod_nette'],
    row['Pct_export_prod_nette'],
    ]


fig = px.pie(
    values=values,
    names=labels,
    title="Rpartition 2022 de la production d'lectricit par habitant en %"
)

fig.write_image("image.png")
files.download('image.png')
fig.show()
```

```python
row = df_hab_2022.iloc[0]


labels = ['Thermique', 'Nuclaire', 'olien', 'Solaire', 'Hydraulique', 'Bionergies', 'Import']
values = [
    row["Pct_therm_conso"],
    row['Pct_nucl_conso'],
    row['Pct_eol_conso'],
    row['Pct_sol_conso'],
    row['Pct_hydro_conso'],
    row['Pct_bioener_conso'],
    row["Pct_import_prod_nette"]
    ]

color_discrete_map = {
    'Nuclaire': 'purple',
    'Thermique': 'red',
    'Hydraulique': 'blue',
    'olien': 'green',
    'Solaire': 'yellow',
    'Import': 'orange',
    'Bionergies': 'pink'
}

fig = px.pie(
    values=values,
    names=labels,
    title="Rpartition 2022 de la production d'lectricit par habitant en %",
    color=labels,
    color_discrete_map=color_discrete_map
)
fig.update_yaxes(tickfont_family="Arial Black", tickfont_size=18)
fig.update_layout(title_x=0.5)
fig.write_image("image.png")
files.download('image.png')
fig.show()
```

```python
row = df_hab_2013.iloc[0]


labels = ['Thermique', 'Nuclaire', 'olien', 'Solaire', 'Hydraulique', 'Bionergies', 'Import']
values = [
    row["Pct_therm_conso"],
    row['Pct_nucl_conso'],
    row['Pct_eol_conso'],
    row['Pct_sol_conso'],
    row['Pct_hydro_conso'],
    row['Pct_bioener_conso'],
    row["Pct_import_prod_nette"]
    ]

color_discrete_map = {
    'Nuclaire': 'purple',
    'Thermique': 'red',
    'Hydraulique': 'blue',
    'olien': 'green',
    'Solaire': 'yellow',
    'Import': 'orange',
    'Bionergies': 'pink'
}

fig = px.pie(
    values=values,
    names=labels,
    title="Rpartition 2013 de la production d'lectricit par habitant en %",
    color=labels,
    color_discrete_map=color_discrete_map
)
fig.update_yaxes(tickfont_family="Arial Black", tickfont_size=18)
fig.update_layout(title_x=0.5, legend=dict(
        font=dict(
            size=18),
        yanchor="middle",
        y=0.5))
fig.write_image("image.png")
files.download('image.png')
fig.show()
```
