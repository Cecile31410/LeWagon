```python
from google.colab import auth
import pandas as pd
auth.authenticate_user()
query = "SELECT * FROM team-elec.Energies_renouvelables.Prod_regionale_13_23"
ENR_df = pd.read_gbq(query, project_id = "team-elec")
```

```python
ENR_df
```

```python
ENR_df_yearly=ENR_df.groupby("annee").sum().reset_index()
ENR_df_yearly=ENR_df_yearly.drop(columns=['Region', 'geo_shape_region', 'geo_point_region'])
ENR_df_yearly
```

```python
import kaleido
import os
from google.colab import files
import plotly.express as px
```

```python
df_melted = ENR_df_yearly.melt(id_vars=["annee", "production_totale_renouvelable"],
                               value_vars=["Hydraulique", "Bioenergies", "Eolienne", "solaire", "Gaz"],
                               var_name="Type_Energie", value_name="Valeur")
```

```python
!pip install plotly>=4.0.0
!wget https://github.com/plotly/orca/releases/download/v1.2.1/orca-1.2.1-x86_64.AppImage -O /usr/local/bin/orca
!chmod +x /usr/local/bin/orca
!apt-get install xvfb libgtk2.0-0 libgconf-2-4
```

```python
fig = px.scatter(df_melted, x="annee", y="Valeur", color="Type_Energie", size="Valeur", size_max=55,
                 title="Production d'énergie renouvelable par type 2013-2023", labels={"Valeur": "Production en GWh", "annee": "Année", "Type_Energie": "Type d'énergie"})
fig.update_layout(
    title={
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    title_font=dict(
        family="Arial, sans-serif",
        size=24,
        color="Black"
    )
)
fig.write_image("image.png")
files.download('image.png')
fig.show()
```
