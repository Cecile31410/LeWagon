```python
# Calcul du nombre de valeurs nulles par colonne
null_counts = conso2_df.isnull().sum()
print("\nNombre de valeurs nulles par colonne :")
print(null_counts)
```

```python
# Attempt to convert to numeric, non-numeric values will become NaN
conso2_df['numeric_column'] = pd.to_numeric(conso2_df["Eolien (MW)"], errors='coerce')

# Identify non-numeric values
non_numeric_values = conso2_df[conso2_df['numeric_column'].isna()]["Eolien (MW)"]

# Print non-numeric values
print("Valeurs non numériques dans la colonne 'Eolien':")
print(non_numeric_values)
print(non_numeric_values.unique())
```

```python
# Remplacer "ND" et "-" par des chaînes vides dans la colonne "Eolien (MW)"
conso2_df["Eolien (MW)"].replace(["ND", "-"], "", inplace=True)
```

```python
conso2_df["Eolien (MW)"] = pd.to_numeric(conso2_df["Eolien (MW)"], errors='coerce')
```

```python
conso2_df.info()
```

```python
conso2_df['Date - Heure'] = pd.to_datetime(conso2_df['Date - Heure'], errors='coerce')
print(conso2_df['Date - Heure'].dtype)
```

```python
conso2_df['time_bins'] = conso2_df['Date - Heure'].dt.floor('2h')
grouped_conso2h_df=conso2_df.groupby(['time_bins', 'Région'])[['Consommation (MW)', 'Thermique (MW)', 'Nucléaire (MW)', 'Eolien (MW)', 'Solaire (MW)', 'Hydraulique (MW)', 'Bioénergies (MW)']].sum().reset_index()
grouped_conso2h_df.head()
```

```python
grouped_conso2h_df.tail()
```

```python
grouped_conso2h_df.to_csv("Conso_région_group2h.csv", index=False)
```
